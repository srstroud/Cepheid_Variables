```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

## Recreating the 1912 Figures

```python
leavitt = np.loadtxt('1912_Periods.txt', skiprows=1)
lea_harvard = leavitt[:,0]
lea_max = leavitt[:,1]
lea_min = leavitt[:,2]
lea_per = leavitt[:,3]
lea_pern = np.sort(lea_per)
ind = np.argsort(lea_per)
lea_maxn = lea_max[ind]
lea_minn = lea_min[ind]
```

```python
plt.plot(lea_pern, lea_maxn, 'b.-')
plt.plot(lea_pern, lea_minn, 'b.-')
plt.gca().invert_yaxis()
plt.title('Circular 173, Fig 1')
plt.xlabel('Period')
plt.ylabel('Brightness')
plt.show()
```

```python
x = np.log10(lea_pern)
y1 = lea_maxn
y2 = lea_minn
z1 = np.polyfit(x, y1, 1)
p1 = np.poly1d(z1)
z2 = np.polyfit(x, y2, 1)
p2 = np.poly1d(z2)

plt.plot(x, y1, 'bo-')
plt.plot(x, y2, 'bo-')
plt.plot(x, p1(x),"b--")
plt.plot(x, p2(x),"b--")
plt.gca().invert_yaxis()

plt.title('Circular 173, Fig 2')
plt.xlabel('log(Period)')
plt.ylabel('Brightness')
#plt.xlim(0, 2.2)
#plt.ylim(16.5, 11)

plt.show()
```

## Tuscana Cepheids in GCVS 5.1

```python
gcvs = np.loadtxt('gcvs5_periods.txt', skiprows=2, usecols=(0,4,5,10), delimiter='|')
gcvs_number = gcvs[:,0]
gcvs_max = gcvs[:,1]
gcvs_min = gcvs[:,2]
gcvs_per = gcvs[:,3]

gcvs_periodn = np.sort(gcvs_per)
ind = np.argsort(gcvs_per)
gcvs_maxn = gcvs_max[ind]
gcvs_minn = gcvs_min[ind]
```

```python
plt.plot(lea_pern, lea_maxn, 'b.-')
plt.plot(lea_pern, lea_minn, 'b.-')
plt.plot(gcvs_periodn, gcvs_maxn, 'r.-')
plt.plot(gcvs_periodn, gcvs_minn, 'r.-')
plt.gca().invert_yaxis()

plt.title('Historical and Modern SMC Cepheids')
plt.xlabel('Period')
plt.ylabel('Brightness')
plt.show()
```

```python
%matplotlib notebook
x = np.log10(lea_pern)
x2 = np.log10(gcvs_periodn)
y1 = lea_maxn
y2 = lea_minn
y3 = gcvs_maxn
y4 = gcvs_minn

z1 = np.polyfit(x, y1, 1)
p1 = np.poly1d(z1)
z2 = np.polyfit(x, y2, 1)
p2 = np.poly1d(z2)

z3 = np.polyfit(x2, y3, 1)
p3 = np.poly1d(z3)
z4 = np.polyfit(x2, y4, 1)
p4 = np.poly1d(z4)


plt.plot(x, y1, 'b.-')
plt.plot(x, y2, 'b.-')
plt.plot(x2, y3, 'r.-')
plt.plot(x2, y4, 'r.-')

plt.plot(x, p1(x),"b--")
plt.plot(x, p2(x),"b--")
plt.plot(x2, p3(x2),"r--")
plt.plot(x2, p4(x2),"r--")
plt.gca().invert_yaxis()

plt.title('Historical and Modern SMC Cepheids')
plt.xlabel('log(Period)')
plt.ylabel('Brightness')

plt.show()
```
